```python
from crewai import Agent,Task,Crew,Process,LLM
import os 
import torch
```

```python
from dotenv import load_dotenv
load_dotenv(dotenv_path="./app.env")
gemini_api_key = os.getenv("GEMINI_API_KEY")
```

```python
llm = LLM(
    model="gemini/gemini-2.0-flash",
    temperature=0,
)
```

```python
auther_agent = Agent(
    role="author", ## it is ajent job accountent-teacher-
    goal="generate stores given some details about it ",
    backstory="egyption auther with five years experiance in stores and inspre of  shekspear" ,#write task how to  make it
    llm=llm ,
    verbose=True
)
```

```python
task =Task(
     description=
       '\n'.join([
        "generate story about {person_name} and  subject is {subject}",
    ]),
    expected_output= '\n'.join([
        "maximum ten lines of story"

    ]),
    agent=auther_agent
        

)
```

```python
summerizer_agent = Agent(
    role="summerizer_author", 
    goal="summrize storey and keep import information ",
    backstory="you ar an experianced sstory summrizer with 10 yeears experiance ",
    llm=llm ,
    verbose=True
)
Summrizer_task = Task(
     description=
       '\n'.join([
           "summrize storey and keep import information"
        ,
    ]),
    expected_output= '\n'.join([
        "summrized story"

    ]),
    agent=summerizer_agent


)
```

```python
#ctrear Crew 
crew = Crew(
    agents=  [
      auther_agent,
      summerizer_agent
    ],
    tasks=[
       task,
       Summrizer_task
    ],
    process=Process.sequential

)
```

```python
crew.kickoff(
    inputs={
        "person_name" : "ahmed Gaber" , 
        "subject" : "his goal in ai "
    }
)
```

```python
writer_agent =Agent(
        role="Writer", 
    goal= "\n".join([
        " write profitional articles and good content in details ",
        " articls must cover all points  for given project name  ",
        " please article must be at least 10 Pges ",
        "article must be well-structured  contain titles  and sub titles nad "
    ]),
    backstory="\n".join([
        "your  artcles is famous and well-structured ",
        "you are Professional writer",
        "you have 10 years experiance in this field ",
        "you awarded more prizes in this field due to your proffessional content",
        "you are desigent to help in createing articles ",


    ]),
    llm=llm ,
    verbose=True

    
)
Writer_task = Task(
    description="\n".join([
        " your task is to genrate well-structured articls about given {project_name} ",
        "the task is to show all information in details for given {project_name}",
        " please divide aricls into  titles and subtitles ,introduction , goal, focal points  and please add another points as much you can to make it prrofinal ",
        "explaining the project in detail"
    ]),
    expected_output="well-structured articls",
    agent=writer_agent

)
```

```python
summry_agent =Agent(
        role="summerizer", 
    goal= "\n".join([
        "summrize give article in short summrized lines covers all article points "
    ]),
    backstory="\n".join([
        "tou are professional article summrizer ",
        "you did task more and more before it is easy task for you",
        "you have 10 years experiance in this field ",
        "you awarded more prizes in this field due to your proffessional summrization",
        "you are desigent to help in summrizing  articles ",


    ]),
    llm=llm ,
    verbose=True

    
)
summry_task = Task(
    description="\n".join([
        " your task is to generate  short summaries for given article  ",
        " please beforee summrization do below steps :"
        " 1- read article carfully  "
        " 2- start define all paraghraphes and its title and  subtitles  ",
        " 3-  start summrization  in short and please output of summrization must contain  bullet points, and TL, DR sections",
    ]),
    expected_output=" Generated short summaries about given article ",
    agent=summry_agent

)
```

```python
factChecker_agent =Agent(
        role="content Checker", 
    goal= "\n".join([
        "Verify technical accuracy of the article content "
    ]),
    backstory="\n".join([
        "tou are professional article content Checker ",
        "you did task more and more before it is easy task for you",
        "you have 10 years experiance in this field ",
        "you awarded more prizes in this field ",
        "you are desigent to help in article  content Checker ",


    ]),
    llm=llm ,
    verbose=True

    
)
factChecker_task = Task(
    description="\n".join([
        " your task is to Verifi technical accuracy of the content  ",
        "output must be report "
        " report must be well strucred"
        "genrated output report must contain  report title  as Content Evaluation Reoport   ,"
        " genrated report must contain Conclusion about content accuarcy  in one section in few lines  "
        " creat new section  for report result and it must include   good or verygood or exellent, or bad content",
        " the output must be have percentage of accuracy  ",
        " output must contain how to imprive accuracy  in new section named how to improve contetent"
        ,
    ]),
    expected_output=" report about content ",
    agent=factChecker_agent

)
```

```python
metadata_agent =Agent(
        role="SEO checker", 
    goal= "\n".join([
        "  Creates SEO-friendly titles, tags, and structured metadata for publishin",
    
    ]),
    backstory="\n".join([
        "tou are professional SEO(Search Engine Optimization) ",
        "you did task more and more before it is easy task for you",
        "you have 10 years experiance in this field ",
        "you awarded more prizes in this field ",
        "you are desigent to help in article to check  Search Engine Optimization ",


    ]),
    llm=llm ,
    verbose=True

    
)
metadata_task = Task(
    description="\n".join([
        " Creates SEO-friendly titles, tags, and structured metadata for publishint  ",
        "please follow the steps below "
        " 1- read article carfully  "
        " 2- start define all meta data required for SEO  ",
        " 3- start define all metadata like Keyword research,Content creation,Link building,Mobile optimization,Structured data markup  ",
        " provided examples in point 3  is an example and for helping you , you can use it if need needed ",
        " after enhancment , please  regenrate the article with new spects add to it for SEO"
        ,
    ]),
    expected_output="new genrated article after check SEO and ",
    agent=metadata_agent

)
```

```python
crew = Crew(
    agents= [
        writer_agent,
        summry_agent,
        factChecker_agent,
        metadata_agent

    ],
    tasks=[
        Writer_task,
        summry_task,
        factChecker_task,
        metadata_task
    ]
    ,
    process=Process.sequential

)
```

```python
crew.kickoff(
    inputs={
        "project_name" : "role of AI in daily Jobs" , 
    }
)
```
